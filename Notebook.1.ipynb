```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Análisis Exploratorio de Datos (EDA)

```python
# Cargar los datos
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')


```

```python
train_df.info()
```

```python
test_df.info()
```

```python
# Exploración inicial
print("Dimensiones del conjunto de entrenamiento:", train_df.shape)
print("Dimensiones del conjunto de prueba:", test_df.shape)
print(train_df.head())
print(train_df.describe())
```

```python
print(train_df.info())
```

```python
# Visualización de datos
sns.countplot(x='Survived', data=train_df)
plt.show()
```

```python
# Mostrar distribución de variables categóricas importantes (Sex, Pclass, Embarked)
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
sns.countplot(x='Sex', data=train_df)
plt.title('Distribución de Sex')
```

```python

plt.subplot(1, 3, 2)
sns.countplot(x='Pclass', data=train_df)
plt.title('Distribución de Pclass')
```

```python
plt.subplot(1, 3, 3)
sns.countplot(x='Embarked', data=train_df)
plt.title('Distribución de Embarked')
plt.show()
```

```python
# Concatenar los conjuntos de datos de entrenamiento y prueba para un análisis más completo
data_df = pd.concat([train_df, test_df], ignore_index=True)



```

```python
# Explorar las primeras filas del DataFrame
print(data_df.head())
```

```python

# Resumen estadístico de variables numéricas
print(data_df.describe())
```

```python

# Verificar valores nulos
print(data_df.isnull().sum())
```

```python
# Visualización de la distribución de la variable objetivo "Survived"
sns.countplot(x='Survived', data=train_df)
plt.title('Distribución de Sobrevivientes')
plt.show()
```

```python
# Distribución de clases en la variable 'Pclass'
sns.countplot(x='Pclass', data=data_df)
plt.title('Distribución de Clases')
plt.show()
```

```python
# Distribución de edades
sns.histplot(data=data_df, x='Age', bins=20, kde=True)
plt.title('Distribución de Edades')
plt.show()
```

```python
# Distribución de género
sns.countplot(x='Sex', data=data_df)
plt.title('Distribución de Género')
plt.show()
```

```python
# Distribución de tarifas pagadas
sns.histplot(data=data_df, x='Fare', bins=30, kde=True)
plt.title('Distribución de Tarifas')
plt.show()
```

```python
numeric_cols = data_df.select_dtypes(include=np.number)
corr_matrix = numeric_cols.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```
